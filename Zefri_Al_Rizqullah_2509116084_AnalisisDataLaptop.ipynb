## **Outliers Values**


### **Pengecekan Outlier dengan Metode IQR**


Metode IQR (Interquartile Range) digunakan untuk mendeteksi nilai yang berada jauh di luar
distribusi normal data. Batas outlier ditentukan sebagai: `< Q1 - 1.5×IQR` atau `> Q3 + 1.5×IQR`.


````python
# Pengecekan outlier menggunakan metode IQR
num_cols = df.select_dtypes(include=['int64', 'float64']).columns
results = []

for col in num_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    n_outlier = ((df[col] < lower) | (df[col] > upper)).sum()
    pct = n_outlier / len(df) * 100
    results.append({
        'Kolom': col, 'Q1': round(Q1, 2), 'Q3': round(Q3, 2),
        'IQR': round(IQR, 2), 'Lower Bound': round(lower, 2),
        'Upper Bound': round(upper, 2), 'Jumlah Outlier': n_outlier,
        'Persentase (%)': round(pct, 2)
    })

out_df = pd.DataFrame(results)
print("=== HASIL PENGECEKAN OUTLIER (METODE IQR) ===")
print(out_df.to_string(index=False))
````

# **Construct Data**


Construct Data adalah tahap pembuatan fitur baru (*feature engineering*) untuk memperkaya dataset
dan meningkatkan kemampuan analisis. Fitur baru dibuat berdasarkan transformasi atau kombinasi
kolom yang sudah ada.


## **1. Transformasi Harga: `Price_actual`**


Kolom `Price` dalam dataset berada dalam skala **logaritmik natural** (ln). Transformasi balik
menggunakan fungsi eksponensial (`exp`) diperlukan untuk mendapatkan nilai harga aktual yang
dapat diinterpretasikan secara bisnis.

$$\text{Price\_actual} = e^{\text{Price}}$$


````python
# Membuat kolom harga aktual (transformasi balik dari ln-price)
df['Price_actual'] = np.exp(df['Price'])

print("=== TRANSFORMASI HARGA ===")
print(f"Price (ln) range  : {df['Price'].min():.3f} – {df['Price'].max():.3f}")
print(f"Price (aktual) min: USD {df['Price_actual'].min():,.0f}")
print(f"Price (aktual) max: USD {df['Price_actual'].max():,.0f}")
print(f"Price (aktual) rata-rata : USD {df['Price_actual'].mean():,.0f}")
print(f"Price (aktual) median    : USD {df['Price_actual'].median():,.0f}")
print()
print("Kolom 'Price_actual' berhasil dibuat!")
````

## **2. Kategorisasi Storage: `Storage_type`**


Dibuat kolom baru `Storage_type` yang mengkategorikan jenis penyimpanan laptop berdasarkan
kombinasi nilai HDD dan SSD:
- **SSD Only**: Hanya menggunakan SSD (HDD = 0, SSD > 0)
- **HDD Only**: Hanya menggunakan HDD (HDD > 0, SSD = 0)
- **Both**: Menggunakan keduanya (HDD > 0 dan SSD > 0)
- **Neither**: Tidak ada keduanya (kemungkinan eMMC atau data tidak tercatat)


````python
# Membuat kolom kategori jenis penyimpanan
def categorize_storage(row):
    if row['HDD'] > 0 and row['SSD'] > 0:
        return 'Both (HDD+SSD)'
    elif row['SSD'] > 0:
        return 'SSD Only'
    elif row['HDD'] > 0:
        return 'HDD Only'
    else:
        return 'Neither'

df['Storage_type'] = df.apply(categorize_storage, axis=1)

print("=== DISTRIBUSI JENIS PENYIMPANAN (Storage_type) ===")
st_dist = df['Storage_type'].value_counts()
for label, count in st_dist.items():
    print(f"  {label:<20}: {count:4d} laptop ({count/len(df)*100:.1f}%)")
print(f"\nKolom 'Storage_type' berhasil dibuat!")
````

## **3. Kategorisasi Segmen Harga: `Price_segment`**


Dibuat kolom `Price_segment` untuk mengelompokkan laptop ke dalam segmen harga pasar:
- **Entry Level**: Harga < USD 25.000
- **Mid Range**: USD 25.000 – 60.000
- **Upper Mid**: USD 60.000 – 100.000
- **Premium**: Harga > USD 100.000


````python
# Membuat kolom segmen harga
def categorize_price(price):
    if price < 25000:
        return 'Entry Level'
    elif price < 60000:
        return 'Mid Range'
    elif price < 100000:
        return 'Upper Mid'
    else:
        return 'Premium'

df['Price_segment'] = df['Price_actual'].apply(categorize_price)

# Urutan segmen
seg_order = ['Entry Level', 'Mid Range', 'Upper Mid', 'Premium']
print("=== DISTRIBUSI SEGMEN HARGA ===")
seg_dist = df['Price_segment'].value_counts().reindex(seg_order)
for seg, count in seg_dist.items():
    print(f"  {seg:<15}: {count:4d} laptop ({count/len(df)*100:.1f}%)")
print(f"\nKolom 'Price_segment' berhasil dibuat!")
````

# **Data Reduction**


Data Reduction bertujuan untuk mengurangi dimensi atau volume data tanpa kehilangan informasi
esensial. Dilakukan melalui analisis korelasi untuk mengidentifikasi fitur yang paling relevan
terhadap variabel target (harga laptop).


## **Analisis Korelasi Pearson terhadap Harga**


````python
# Analisis korelasi fitur numerik terhadap Price
num_features = ['Ram', 'Weight', 'Ppi', 'HDD', 'SSD', 'TouchScreen', 'Ips']
corr_series = df[num_features + ['Price']].corr()['Price'].drop('Price').sort_values(ascending=False)

print("=== KORELASI PEARSON TERHADAP HARGA (Price) ===\n")
print(f"{'Fitur':<20} {'Korelasi':>10}  {'Interpretasi':<25} {'Prioritas'}")
print("─" * 75)

for feat, corr_val in corr_series.items():
    if abs(corr_val) >= 0.5:
        interp, priority = 'Kuat', 'Tinggi'
    elif abs(corr_val) >= 0.3:
        interp, priority = 'Sedang', 'Sedang'
    elif abs(corr_val) >= 0.15:
        interp, priority = 'Lemah', 'Rendah'
    else:
        interp, priority = 'Sangat Lemah', 'Sangat Rendah'
    arah = '(+)' if corr_val >= 0 else '(-)'
    print(f"  {feat:<18} {corr_val:>10.3f}  {interp+' '+arah:<25} {priority}")
````

## **Visualisasi Korelasi**


````python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))
fig.suptitle('Analisis Korelasi Fitur terhadap Harga Laptop',
             fontsize=13, fontweight='bold')

# Bar plot korelasi
colors_bar = ['#E74C3C' if abs(v) >= 0.5 else '#E67E22' if abs(v) >= 0.3
              else '#27AE60' if abs(v) >= 0.15 else '#BDC3C7'
              for v in corr_series.values]
bars = ax1.barh(corr_series.index[::-1], corr_series.values[::-1],
                color=colors_bar[::-1], edgecolor='white', linewidth=0.8)
ax1.axvline(x=0, color='black', linewidth=0.8)
for bar, val in zip(bars, corr_series.values[::-1]):
    x_pos = bar.get_width() + 0.01 if val >= 0 else bar.get_width() - 0.01
    ha = 'left' if val >= 0 else 'right'
    ax1.text(x_pos, bar.get_y() + bar.get_height()/2, f'{val:.3f}',
             va='center', ha=ha, fontsize=9, fontweight='bold')
ax1.set_title('Korelasi terhadap Price', fontweight='bold')
ax1.set_xlabel('Nilai Korelasi Pearson')
ax1.grid(axis='x', alpha=0.3)
ax1.spines['top'].set_visible(False)
ax1.spines['right'].set_visible(False)

# Heatmap matriks korelasi
all_num = df[['Ram','Weight','Price','Ppi','HDD','SSD','TouchScreen','Ips']].corr()
mask = np.zeros_like(all_num, dtype=bool)
mask[np.triu_indices_from(mask, k=1)] = True
sns.heatmap(all_num, annot=True, fmt='.2f', cmap='RdYlBu_r', center=0,
            vmin=-1, vmax=1, ax=ax2, linewidths=0.5, annot_kws={'size': 8},
            mask=False)
ax2.set_title('Matriks Korelasi Lengkap (Heatmap)', fontweight='bold')
ax2.tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
````
